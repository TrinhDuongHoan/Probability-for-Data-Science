### Lab 2. Quy tắc tính xác suất. Xác suất hình học (ngày 15/01)

* $E^c = \Omega \setminus E$: biến cố **đối** (complement) của $E$, biến cố "$E$ không xảy ra"

* $E \cup F$: biến cố $E$ **hoặc** (or) $F$, biến cố "$E$ xảy ra hoặc $F$ xảy ra"

* $E \cap F$: biến cố $E$ **và** (and) $F$, biến cố "$E$ xảy ra và $F$ xảy ra"

* $E \setminus F$: biến cố $E$ **không** (not) $F$, biến cố "$E$ xảy ra nhưng $F$ không xảy ra"

* $E \subset F$: $E$ **kéo theo** (imply) $F$, $E$ xảy ra thì $F$ xảy ra

* $E = F$: $E$ **là** (is) $F$, $E$ xảy ra khi và chỉ khi $F$ xảy ra

* $E \cap F = \emptyset$: $E, F$ **rời nhau** (disjoint) hay **xung khắc** (mutually exclusive), $E, F$ không thể đồng thời xảy ra

### Các yêu cầu (tiên đề) của xác suất

Xét thí nghiệm $T$ với không gian mẫu $\Omega$, một hàm $P$ gắn mỗi biến cố $E \subset \Omega$ với số thực $P(E)$ được gọi là một **độ đo xác suất** (probability measure) trên $\Omega$ nếu $P$ thỏa mãn 3 tiên đề:

1. Với mọi biến cố $E \subset \Omega, 0 \le P(E) \le 1$

2. Với mọi dãy biến cố $E_1, E_2, ..., E_k$ đôi một xung khắc $\left(E_i \cap E_j = \emptyset, \forall i \ne j \right)$:$$P\left(\bigcup_{i=1}^{k}E_i\right) = \sum_{i=1}^{k}P(E_i)$$

3. $P(\Omega) = 1$

### Các tính chất (hệ quả) cơ bản của xác suất

### Tóm lược lý thuyết

* $P(E^c) = 1 - P(E)$

* $P(\emptyset) = 0$

* Nếu $E_1 \subset E_2$ thì $P(E_1) \le P(E_2)$ và $P(E_2 \setminus E_1) = P(E_2) - P(E_1)$

* Nếu $E_1 \cap E_2 = \emptyset$ thì $P(E_1 \cup E_2) = P(E_1) + P(E_2)$

* $P(E_1 \cup E_2) = P(E_1) + P(E_2) - P(E_1 \cap E_2)$ (**addition law of probability**)

* $P(E_1 \cup E_2 \cup E_3) = P(E_1) + P(E_2) + P(E_3)\\ - P(E_1 \cap E_2) - P(E_1 \cap E_3) - P(E_2 \cap E_3)\\ + P(E_1 \cap E_2 \cap E_3)$

* $P\left(\bigcup_{i=1}^{\infty}E_i\right) \le \sum_{i=1}^{\infty}P(E_i)$ (**union bound**)

* $P\left(\bigcap_{i=1}^{\infty}E_i\right) \ge 1 - \sum_{i=1}^{\infty}P(E_i^c)$ (**Bonferroni inequality**)

### Độc lập xác suất

2 biến cố $\{E, F\}$ được gọi là **độc lập** (independent) nhau nếu:
$$P(EF) = P(E)P(F)$$
Họ các biến cố $\{A_1, A_2, ..., A_n\}$ được gọi là độc lập nếu với mọi tập con khác rỗng, hữu hạn $\{B_1, B_2, ..., B_k\}$ của họ, ta có:
$$P(B_1 B_2  ... B_k) = P(B_1)P(B_2)...P(B_k)$$

**Mô hình xác suất "lặp lại thí nghiệm độc lập"**: thực hiện lặp lại thí nghiệm $T$ nhiều lần một cách độc lập, gọi $A_i$ là biến cố "liên quan đến lần thực hiện thứ $i$" thì ta có công thức tính xác suất:
$$P(A_1 A_2  ... A_n) = P(A_1)P(A_2) ... P(A_n)$$

```python
#Simulate việc ném phi tiêu ở trên
import math

# Cài đặt hàm ném phi tiêu này sai (không ngẫu nhiên theo bán kính)
def nem_phi_tieu_1(): # hinh tron C co ban kinh 1
    r = random.random()
    phi = random.uniform(0, 360)
    
    return (r*math.cos(phi), r*math.sin(phi))

# Cài đặt hàm ném phi tiêu này đúng (ngẫu nhiên theo điểm trong hình tròn C)
def nem_phi_tieu(): # hinh tron C co ban kinh 1
    while True:
        x, y = 2*random.random() - 1, 2*random.random() - 1
        r = math.sqrt(x**2 + y**2)
        if r <= 1:
            return (x, y)

def tan_suat_trung(t, N): # hinh tron D co ban kinh t
    return sum(math.sqrt(x**2 + y**2) <= t for x, y in (nem_phi_tieu() for _ in range(N))) / N
```

```python
tan_suat_trung(1/2, 100_000)
```

### Bài tập 4. 
You get a stick and break it randomly into three pieces. What is the probability that you can make a triangle using the three pieces? You can assume the break points are chosen completely at random, i.e. if the length of the original stick is $1$ unit, and $x,y,z$ are the lengths of the three pieces, then $(x,y,z)$ are uniformly chosen from the set
    \begin{equation*}\{(x,y,z)\in\mathbb{R}^{3}|x+y+z=1,x,y,z\geq0\}.
    \end{equation*}

This is again a problem on a continuous probability space. The basic idea is pretty simple. First, we need to identify the sample space $S$. In this case the sample space is going to be a two-dimensional set. Second, we need to identify the set $A$ that contains the favorable outcomes (the set of $(x,y,z)$ in $S$ that form a triangle). And finally, since the space is uniform, we will divide area of set $A$ by the area of $S$ to obtain $P(A)$.

First, we need to find the sets $S$ and $A$. This is basically a geometry problem. The two sets, $S$ and $A$, are shown in the figure.
    Note that in $\mathbb{R}^{3},\ \ x+y+z=1$ represents a plane that goes through the points $(1,0,0),(0,1,0),(0,0,1)$. To find the sample space $S$, note that $S=\{(x,y,z)\in\mathbb{R}^{3}|x+y+z=1,x,y,z\geq0\}$, thus $S$ is the part of the plane that is shown in figure.
    To find the set $A$, note that we need $(x,y,z)$ to satisfy the triangle inequality
    \begin{equation*}
        \begin{array}{c}{{x+y> z,}}\\ {{y+z> x,}}\\ {{x+z> y.}}\end{array}
    \end{equation*}
    Note that since $x+y+z=1$, we can equivalently write the three equations as
    \begin{equation*}
        \begin{array}{r}
        {x< {\frac{1}{2}},}\\ \\
        {y< {\frac{1}{2}},}\\ \\
        {z< {\frac{1}{2}.}}
        \end{array}
    \end{equation*}
    Thus, we conclude that the set $A$ is the area shown in Figure 2. In particular, we note that the set $S$ consists of four triangles with equal areas. Therefore, its area is four times the area of $A$, and we have
    \begin{equation*}
        P(A)={\frac{\mathrm{Area~of~}A}{\mathrm{Area~of~}S}}={\frac{1}{4}}.
    \end{equation*}

**Bài 3.** Cho một danh sách xác suất bắn trúng mục tiêu của $n$ xạ thủ là $p_1, p_2, ..., p_n$ với các số $p_i \in (0;1), \, \forall i = 1,2,...,n$. Viết hàm CalculateProb nhận vào danh sách đó, tính và in ra các giá trị sau đây:
- Xác suất để tất cả đều bắn trúng.
- Xác suất để không ai bắn trúng.
- Xác suất để có ít nhất một người bắn trúng.
- Xác suất để có đúng hai người bắn trúng.

```python
def CalculateProb(p):
    tatCaTrung = 1
    khongAiTrung = 1
    dungHaiNguoiTrung = 0
    
    for p_i in p:
        tatCaTrung *= p_i
        khongAiTrung *= (1 - p_i)
    
    for i in range(0, len(p)):
        for j in range(0, i):
            prob = p[i] * p[j]
            for z in range(0, len(p)):
                if z not in [i, j]:
                    prob *= (1 - p[z])
            dungHaiNguoiTrung += prob
    
    print('Xác suất để tất cả đều bắn trúng là', tatCaTrung)
    print('Xác suất để không ai bắn trúng là', khongAiTrung)
    print('Xác suất để có ít nhất một người bắn trúng là', 1 - khongAiTrung)
    print('Xác suất để có đúng hai người bắn trúng là', dungHaiNguoiTrung)
```

```python
p = [random.uniform(0, 1) for _ in range(3)]
CalculateProb(p)
```
